Converting

s2n(s) - packed string to number

n2s(n) - number to packed string

s2b(s) - packed string to binary string

b2s(b) - binary string to packed string

```
import libnum
s="flag{pcat}" 
print libnum.s2n(s)
```

```
import libnum

n=0x666c61677b706361747d 

print libnum.n2s(n)
```

```
import libnum
b='01110000011000110110000101110100' 
print libnum.b2s(b)
```
