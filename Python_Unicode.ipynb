Python的Unicode支援非常的優秀，簡單而且好用，

在Python裡的字串有兩種:
1.    str
2.   unicode


在Python表示unicode字串的方式，是在字串前面加上一個u，

為了讓Python直譯器能正確地解析我們在程式碼裡打的中文字，我們得告訴它我們用的是什麼編碼，因此得加上

```
# -*- coding: utf8 -*-
```



````
# -*- coding: utf8 -*-
msg = u'今天天氣真好'
print msg
print u'字數:', len(msg)
````

````
print u"\u4f60\u9700\u8981\u91cd\u65b0\u767b\u9646"
````

````
# -*- coding: utf8 -*-
msg = u'今天天氣真好'
encoded = msg.encode('utf8')
print repr(encoded)
````

````
# -*- coding: utf8 -*-
encoded = '\xe4\xbb\x8a\xe5\xa4\xa9\xe5\xa4\xa9\xe6\xb0\xa3\xe7\x9c\x9f\xe5\xa5\xbd'
msg = encoded.decode('utf8')
print msg
````
